- **Coeficiente de Gini** Para el nodo $t$ y en cn $J$ clases
$$i_G(t)=\sum^J_{j=1}p(j|t)[1-p(j|t)]$$
$$= 1 - \sum^J_{j=1}p(j|t)^2$$
<br>

- **Entrop√≠a** Para el nodo $t$ y con $J$ clases
$$i_e(t)=-\sum^J_{j=1}p(j|t)log[p(j|t)]$$
